```python
mnist = fetch_openml('mnist_784')
```

```python
x, y = mnist['data'], mnist['target']

```

```python

x_train, x_test = x[:60000], x[60000:70000]

y_train, y_test = y[:60000], y[60000:70000]
```

```python

shuffle_index = np.random.permutation(60000)

x_train, y_train = x_train.to_numpy()[shuffle_index], y_train.to_numpy()[shuffle_index]
```

```python

# Creating a 2-detector
#y_train = y_train.astype(np.int8)
#y_test = y_test.astype(np.int8)
y_train_temp = [i for i in range(10)]
y_test_temp = [i for i in range(10)]
for i in range(10):
    temp_var = str(i)
    y_train_temp[i] = (y_train == temp_var )
    y_test_temp[i] = (y_test == temp_var )
    
#y_train_2 = (y_train == '2')

#y_test_2 = (y_test == '2')
```

```python
# Train a logistic regression classifier
from sklearn.linear_model import LogisticRegression

```

```python
clf = [i for i in range(10)]
for i in range(10):
    clf[i] = LogisticRegression(tol=0.1)
for i in range(10):
    clf[i].fit(x_train,y_train_temp[i])
#clf.fit(x_train, y_train_2)
```

```python

from sklearn.model_selection import cross_val_score
a = [0 for i in range(10)]
for i in range(10):
    a[i] = cross_val_score(clf[i], x_train, y_train_temp[i], cv=3, scoring="accuracy")

#a = cross_val_score(clf[0], x_train, y_train_temp[0], cv=3, scoring="accuracy")
```
